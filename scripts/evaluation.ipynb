```python
# Standard built-in modules
import os
import warnings

# Third-party modules
import matplotlib.pyplot as plt
import torch

# Local modules
from datasets import PCSDataSet

# Configurations
warnings.filterwarnings("ignore")  
root_path = os.path.dirname(os.path.abspath(''))
```

### Data loaders
We first check the performance of our datasets and dataloaders by showing a 
random batch of 6 pre-op and post-op images generated by our train dataloader.

```python
train_dataset = PCSDataSet(mode='train', train_size=1000)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=9,
                                           shuffle=True, num_workers=4, 
                                           pin_memory=True, drop_last=True)
sample_batch = next(iter(train_loader))
```

```python
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
for i, data in enumerate(sample_batch['pre']):
    axes[i//3, i%3].imshow(data.permute(1, 2, 0).numpy(), cmap='gray')
    axes[i//3, i%3].axis('off')
```

```python
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
for i, data in enumerate(sample_batch['post']):
    axes[i//3, i%3].imshow(data.permute(1, 2, 0).numpy(), cmap='gray')
    axes[i//3, i%3].axis('off')
```
